## Implementación de fibonacci recursiva

```python
def fibonacci_recursivo(n: int | str) -> int:
    """
    Función que calcula el n-ésimo número de Fibonacci de forma recursiva.

    args:
        n (int): La posición en la secuencia de Fibonacci.

    returns:
        int: El n-ésimo número de Fibonacci.
    
    """
    if isinstance(n, str):
        try:
            n = int (n)
        except ValueError:
            raise ValueError("El valor de n debe ser un entero no negativo")

    if n < 0:
        raise ValueError("El valor de n debe ser un entero no negativo")
    if (n < 2):
        return n
    else:
        n = fibonacci_recursivo(n-1)+fibonacci_recursivo(n-2)
        return n
```

```python
fibonacci_recursivo(9)
```

# Implementación no recursiva

```python
cache = {0: 0, 1: 1}
```

```python
def fibonacci_no_recursivo(n):
    if isinstance(n, str):
        try:
            n = int (n)
        except ValueError:
            raise ValueError("El valor de n debe ser un entero no negativo")

    if n < 0:
        raise ValueError("El valor de n debe ser un entero no negativo")
    if (n < 2):
        return n
    else:
        for i in range(2, n+1):
            if i not in cache:
                cache[i] = cache[i - 1] + cache[i - 2]


    return cache[n]
```

```python
fibonacci_no_recursivo(456)
```
